# 環境について

Jupyter上で以下の言語を扱うことが出来ます。

- Python
- Java(OpenJDK 11)
- JavaScript
- TypeScript
- C++(11,14,17)

また、sshでログインし、以下のコマンドを実行することで「go言語」も扱うことが出来るようになります。

```
$ /usr/lib/alpha/bin/setup_go_kernel
```

sshログインして C++ をコンパイルする場合、以下を実行することで C++14 が利用できます。

```
$ scl enable devtoolset-7 bash
```

# 注意事項

Jupyterでコーディングしていると、プログラム実行用のプロセスが増えていきます。  
定期的に、以下の画面で `Shutdown` してください。



実行プロセスが多くなった場合は、管理者がプロセスを停止させることがあるため、Jupyter上ではデーモン的なプログラムを動かさないでください。  
sshでログインして通常のシェルから実行しているプロセスに関しては停止されません。

一部の Kernel（JavaやC++）は、同じセルを再実行するとクラス定義が重複してエラーになることがあります。  
その場合は Kernel を Restart してください。
