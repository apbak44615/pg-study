# インタプリタの仕組み

Pythonやシェルなどのような、コンパイル無しに実行できるインタプリタについて説明します。

## コンパイラ言語との違い

コンパイラ言語との大きな違いは、書いたプログラムをコンパイル無しにすぐに実行できるところです。  
なぜすぐに実行できるかというと、インタプリタ言語もプログラム言語ではありますが、そのプログラムが直接動いているわけではなく、インタプリタ言語を動かす別のプログラムが動いているためです。

シェルであれば`bash`や`csh`などのプログラムが動き、Pythonであれば`python`というプログラムが動きます。  
自分で書いたプログラムが直接動いているわけではありません。

`bash`や`python`というプログラムが、自分が書いたプログラムを読み取って解釈しながら動かしています。

## インタプリタの動き

インタプリタは、実行時にプログラムファイルの構文を解析し、それに応じて処理します。  
インタプリタの種類によって大きく以下のような動作の違いがあります。

- 認識できる一命令ずつ解釈して実行していく
- 1つのファイル全体をいったん解析し、その後に実行していく

プログラム的な構文エラーがあった場合、前者はある程度命令を実行した後に構文エラーの行で初めてエラーとなりますが、後者は実行前の解析でエラーが分かります。  
コンパイラ言語であれば、コンパイルできた段階で構文エラーがないことが保証できますが、インタプリタは実行中あるいは実行しようと解析した際に初めて分かります。  
そのため、少し修正して実行したつもりになっていて構文エラーのままリリースされてしまう・・・というのも起こりがちなので、コンパイラ言語以上に注意した方がよいです。

特にシェル系は前者の一命令ずつ認識しながら実行していくタイプが多いため、ある部分を変えて実行して確認したが、実は修正部分は通ってなかったということもあるため注意してください。

このようにインタプリタ言語は実行時に初めて構文解析などが行われるため、一般的にコンパイル言語より遅くなります。  
中には効率化するために、一度構文解析したファイルの内容を実行しやすい内容にキャッシュしておき、ファイルに変更がなければキャッシュを使って実行するようなものもあります。  
インタプリタによっていろいろ効率化はされているため、遅いと決めつけずに使いやすいところで使っていくのが良いでしょう。
